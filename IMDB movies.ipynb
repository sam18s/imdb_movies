# Top Movies Genre

```python
df['Genre'].value_counts()
```

```python
# Assuming df is your DataFrame containing a column 'Genre' with comma-separated values
# Splitting the 'Genre' column by commas and creating a new DataFrame
df_genre = df['Genre'].str.split(',')

# Exploding the list-like column to separate each genre into its own row
df_genre = df_genre.explode()

# Counting the occurrences of each genre after exploding the data
df_genre.value_counts()
```

```python
# Assuming df_genre contains your data
plt.figure(figsize=(16, 9))

# Assuming df_genre is a pandas Series or DataFrame column
genre = df_genre.value_counts().sort_index()

# Create a horizontal bar plot using Seaborn
sns.barplot(y=genre.index, x=genre.values, palette='bright')
plt.title("Movie Genres")
plt.xlabel("Number of movies")
plt.ylabel("Genres")
plt.show()
```

```python
# Create a figure and set its size using figsize
plt.figure(figsize=(16, 20))

# Create a pie chart using the 'genre' data with autopct='%1.1f%%' to display percentages on the chart
plt.pie(genre, autopct='%.1f')

# Set the title of the pie chart
plt.title("Movies by Genre")

# Display the pie chart
plt.show()
```

# Top 10 Actor

```python
# Split the 'Actors' column by commas and remove spaces before or after each name
actr_name = df['Actors'].str.split(",").apply(lambda x: [i.strip() for i in x])

# Explode the list-like column to separate each actor name into its own row
actr_name = actr_name.explode()

# Count the occurrences of each actor name
top_10_actors = actr_name.value_counts().head(10)

# Display only the top 10 actors and their occurrence counts
top_10_actors
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(9, 6))  # Set the figure size
sns.barplot(x=top_10_actors.values, y=top_10_actors.index, palette='viridis')

# Set the labels and title
plt.xlabel('Occurrences')
plt.ylabel('Actor Name')
plt.title('Top 10 Actors by Occurrence Count')

# Show the plot
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Metascore'], df['Revenue (Millions)'], color='deeppink', marker='*', s=100, alpha=0.6)

# Adding labels and title
plt.title('Relationship between Metascore and Revenue')
plt.xlabel('Metascore')
plt.ylabel('Revenue (Millions)')

# Show the plot
plt.show()
```

# Heatmap of Movies Rating over Years

```python
# Replace 'Year' with the appropriate column name if needed
heatmap_data = df.groupby(['Rating', 'Year']).size().unstack().fillna(0)

# Extract top 10 rating by movie count
top_directors = df['Rating'].value_counts().head(10).index

# Filter the heatmap data for top 10 directors
heatmap_data = heatmap_data.loc[top_directors]

plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size

# Create the heatmap
sns.heatmap(heatmap_data)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Number of Movies Rating over Years')

# Show plot
plt.show()
```
